##Feature scaling
Feature Scaling is a technique to standardize the independent features present in the data in a fixed range. It is performed during the data pre-processing to handle highly varying magnitudes or values or units. If feature scaling is not done, then a machine learning algorithm tends to weigh greater values, higher and consider smaller values as the lower values, regardless of the unit of the values.


###Types
1.Standardization (by z-score)<br>
2.Normalization (Min-Max Scaler {-1 to 1}) <br>
3.Unit vector <br>
4.PCA(Principle Analysis Algorithm)

**1.Standardization**

```python
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
```

```python
total_bill=list(df['total_bill'])
total_bill
```

```python
sns.histplot(total_bill)
```

```python
import numpy as np      ##for finding mean and standard deviation for calculating z-score
mean=np.mean(total_bill) 
std=np.std(total_bill)
mean,std
```

```python
normalized_data=[]         ## manualy appyling Z-score formula and appending all values in a variable 
for i in total_bill:
    z_score=(i-mean)/std ## z_score
    normalized_data.append(z_score)
    
print(normalized_data)
```

```python
sns.histplot(normalized_data) ##now see both diagram they are similar they have no change that's why we use standardize for making values small for easy calculation
```

```python
from sklearn.preprocessing import StandardScaler ## but for every column in hard to us for all time so thats why a predefined function by sklearn
df.head()
```

```python
scaler=StandardScaler()  ## initializing the class in a variable
scaler.fit(df[['total_bill','tip']]) ## we use for calculating mean and standard deviation
import pandas as pd
## we use transform for calculating the z-score (transfrom the old data to new one)
pd.DataFrame(scaler.transform(df[['total_bill','tip']]),columns=['total_bill','tips']) 
```

**2.Normalization (Min-Max scaler)**

```python
import seaborn as sns
df=sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
min_max.fit(df[['distance','fare','tip']])
```

```python
min_max.transform(df[['distance','fare','tip']])
```

```python
min_max.fit_transform(df[['distance','fare','tip']]) ## u can perform both fit and transform in one line code fit_transform
```

```python
min_max.transform([[1,3,4]])
```

**3.Unit vector feature scaling**

```python
import seaborn as sns
df=sns.load_dataset('iris')
df.head()
```

```python
from sklearn.preprocessing import normalize
df.columns
```

```python
normalize(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
import pandas as pd
pd.DataFrame(normalize(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]),
             columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```
