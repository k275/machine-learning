```python
def convert_to_kilobytes(sizes):
    converted_sizes = []
    for size in sizes:
        if 'M' in size:
            # Convert from megabytes to kilobytes
            size_in_kb = str(float(size.replace('M', '')) * 1024)
        else:
            # Remove the 'k' from the size 
            size_in_kb =(size.replace('k', ''))
        converted_sizes.append(size_in_kb)
    return converted_sizes
```

```python
df_copy['Size']=convert_to_kilobytes(df_copy['Size'])
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

## Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

What are the Top 5 most installed Apps in Each popular Categories ??

**Does Size of application has any impact on its popularity ??**

```python
# Using violin plot to plot the relation
plt.figure(figsize=(20,10))
sns.violinplot(x=df_copy.Installs, y=df_copy.Size)
plt.xticks(rotation = 90)
plt.show()
```

###Insights
- We can observe from the plot that there is a large impact by the size of the app on the number of installations.<br>
- The tiny white circle in middle of each plot shows the median value of each value of installations.<br>
- Across the plot, we can see the median grows steadliy higher.<br>
- Initially on Install axis, there is a higher number of outliers with respect to the size of the apps.<br>
- As we progress across the Install axis, the number of outliers decreases and the number of installations increases.<br>
- As the number of installations reaches the maximum value, we see that the app size has reached the lowest values, peaking at posiibly, 100 -110 MB.<br>
- **From this we can conclude that bigger the app, lesser the chance for it to be installed**
